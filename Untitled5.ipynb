```python
#bike 데이터를 bike_sum에 카피하기

bike_sum=bike.copy()

# '월' 컬럼 만들기
bike_sum['월'] = bike_sum['대여일자'].dt.month

# 월별 합계 계산
monthly_sum = bike_sum.groupby('월')['대여건수'].sum().reset_index()

plt.figure(figsize=(8,5))
plt.plot(monthly_sum['월'], monthly_sum['대여건수'], marker='o', color='blue')

# 과학적 표기(1e6) OFF
plt.ticklabel_format(style='plain', useOffset=False, axis='y')


plt.title("월별 대여건수 합계")
plt.xlabel("월")
plt.ylabel("대여건수")
plt.xticks(monthly_sum['월'])
plt.grid(True)
plt.show()
```

```python

# 1️⃣ 이동평균 / 표준편차 계산
window = 5  # 심전도 느낌이면 5~7일 추천

bike['mean'] = bike['대여건수'].rolling(window).mean()
bike['std'] = bike['대여건수'].rolling(window).std()

bike['upper'] = bike['mean'] + 2 * bike['std']
bike['lower'] = bike['mean'] - 2 * bike['std']

# 2️⃣ 그래프
plt.figure(figsize=(12, 6))

# 원래 선 (심전도)
sns.lineplot(bike, x='대여일자', y='대여건수',
             label='Daily', alpha=0.6)

# # 중심선 (이동평균)
# sns.lineplot(bike, x='대여일자', y='mean',
#              label='Rolling Mean', linewidth=2)

# 변동성 밴드 (볼린저 영역)
plt.fill_between(
    bike['대여일자'],
    bike['lower'],
    bike['upper'],
    color='steelblue',
    alpha=0.15,
    label='Volatility Band'
)

plt.title("대여건수 시계열 그래프 (심전도 + 변동성 밴드)")
plt.xlabel("대여일자")
plt.ylabel("대여건수")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# 1️⃣ 데이터 복사
bike_day = bike.copy()

# 2️⃣ 요일 컬럼 추가 (영어 요일)
bike_day['요일'] = bike_day['대여일자'].dt.day_name()

# 3️⃣ 영어 요일 → 한글 전체 이름 매핑
eng2kor_full = {
    'Monday': '월요일',
    'Tuesday': '화요일',
    'Wednesday': '수요일',
    'Thursday': '목요일',
    'Friday': '금요일',
    'Saturday': '토요일',
    'Sunday': '일요일'
}
bike_day['요일'] = bike_day['요일'].map(eng2kor_full)

# 4️⃣ 요일 순서 지정
요일순서 = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']

# 5️⃣ 요일별 평균 계산 & 순서 맞추기
weekday_mean = bike_day.groupby('요일')['대여건수'].mean().reindex(요일순서)



# 6️⃣ 바 색상: 평일 파랑, 주말 빨강
colors = ['skyblue']*5 + ['salmon','salmon']

# 7️⃣ matplotlib으로 바 그래프
plt.figure(figsize=(10,5))
bars = plt.bar(weekday_mean.index, weekday_mean.values, color=colors)
# 5️⃣ 막대 위 숫자 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, f'{int(height)}건',
             ha='center', va='bottom', fontsize=10)



plt.title("요일별 평균 대여건수")
plt.xlabel("요일")
plt.ylabel("평균 대여건수")
plt.show()
```

```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
movies = pd.read_csv("ml-latest-small/movies.csv")

data=ratings.merge(movies, on="movieId")

print(data.head())
print("=========")
print(data["userId"].max())
```

```python
print(data.info())
print(data.describe())
```

```python
data_mr=data.copy()

# (YYYY) 형식으로 끝나는 것만 추출
data_mr['year'] = data_mr['title'].str.extract(r'\((\d{4})\)$')
data_mr['year'] = pd.to_numeric(data_mr['year'], errors='coerce')  # 변환 불가 -> NaN

#NaN결측치는 제거

data_mr = data_mr.dropna(subset=['year'])



# 3️⃣ 산점도 그리기
plt.figure(figsize=(12,6))
plt.scatter(data_mr['year'], data_mr['rating'],alpha=0.05,s=15,color='red')
plt.title("연도별 영화 평점 갯수")
plt.xlabel("연도")
plt.ylabel("평점")
plt.grid(True)
plt.show()
```

```python
# 0️⃣ 연도 추출
data_mr['year'] = data_mr['title'].str.extract(r'\((\d{4})\)$')
data_mr['year'] = pd.to_numeric(data_mr['year'], errors='coerce')
data_mr = data_mr.dropna(subset=['year'])

# 1️⃣ 10년 단위로 묶기
data_mr['decade'] = (data_mr['year'] // 10) * 10

# 2️⃣ 10년 단위 + 평점별 개수 집계
count_table = (
    data_mr
    .groupby(['decade', 'rating'])
    .size()
    .reset_index(name='count')
)

# 3️⃣ 피벗 테이블 생성 후 전치(transpose)
pivot = (
    count_table
    .pivot(index='decade', columns='rating', values='count')
    .fillna(0)
    .astype(int)
    .T   # 🔥 여기서 x/y축 반전
)

# 4️⃣ 히트맵
plt.figure(figsize=(14, 8))
ax = sns.heatmap(
    pivot,
    cmap=['white'],   # 🔥 단일 색만 사용
    annot=True,
    fmt='d',
    cbar=False,
    linewidths=0.5,
    linecolor='black',
    vmin=0,
    vmax=1            # 🔥 모든 값이 같은 색으로 처리됨
)

# 🔥 외곽 테두리 강제
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('black')
plt.title('평점 × 10년 단위 연도별 개수 히트맵')
plt.xlabel('연도 (10년 단위)')
plt.ylabel('평점')
plt.tight_layout()
plt.show()
```

```python
data_mr2=data.copy()

# 1️⃣ 복사본 사용
df = data_mr2.copy()


# 2️⃣ 장르 분리 후 explode
df['genres'] = df['genres'].str.split('|')
df_expanded = df.explode('genres')

# 3️⃣ 장르별 평점 수집
# 필요한 장르만 선택 가능, 예: 'Comedy', 'Action', 'Drama', 'Thriller'
selected_genres = ['Comedy', 'Action', 'Drama', 'Thriller']
df_plot = df_expanded[df_expanded['genres'].isin(selected_genres)]

# 4️⃣ 바이올린 플롯
plt.figure(figsize=(12,6))
sns.violinplot(x='genres', y='rating', data=df_plot, inner='box', palette='Pastel1')

plt.title('장르별 평점 분포 바이올린 플롯')
plt.xlabel('장르')
plt.ylabel('평점')
plt.show()
```
